Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

```python
import math as m
x1,y1= map(int,input("Enter First Quadrant:").split(","))
x2,y2= map(int,input("Enter Second Quadrant:").split(","))

print(f"A({x1},{y1}) B({x2},{y2})")

dist=m.sqrt((x2-x1)**2 + (y2-y1)**2)

print(f"Distance b/w A({x1},{y1}) and  B({x2},{y2}) is: {dist :.2f}")
```

Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.
For example: Input: heads -> 4 legs -> 12
Output: dogs -> 2 chicken -> 2

```python
h,l=map(int,input("Enter total number of heads and legs:").split(","))
#Let c = number of chickens, d = number of dogs
#Eq1:c+d=heads
#Eq2:2c+4d=legs
#solve equation we get:
#d=(legs-2heads)/2
#c=heads-d

d=(l-2*h)//2
c=h-d
print(f"Dogs={d} and Chicken={c}")
```

Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.
Input:
Dimensions of the milk tank
H = 20cm, L = 20cm, B = 20cm

Dimensions of the glass
h = 3cm, r = 1cm

```python
import math as m
H,W,B=map(int,input("Enter Height,Width and Breadth of milk tank in cm:").split(","))
H1,R=map(int,input("Enter Height,radius of glass in cm:").split(","))

t_vol=H*B*W
g_vol=m.pi*H1*(R**2)

total_glasses=t_vol/g_vol
whole_glasses_s=t_vol//g_vol
remain=total_glasses-whole_glasses_s
print(f"No of GLasses:{whole_glasses_s} glasses and {remain:.2f} of a glass")
```
