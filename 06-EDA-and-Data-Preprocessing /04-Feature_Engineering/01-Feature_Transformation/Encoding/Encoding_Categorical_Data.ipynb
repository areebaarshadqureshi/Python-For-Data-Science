## Feature Engineering

Feature engineering is the process of transforming raw data into meaningful input features to improve machine learning model performance, training efficiency, and accuracy.

## Types of FeatureEngineering

- Feature Transformation
- Feature Construction
- Feature Selection
- Feature Extraction

### 1.Feature Transformation
Feature transformation modifies existing features to make them more suitable for algorithms. Key steps include:

- <b>Handling Missing Values:</b> Most Scikit-learn algorithms cannot handle NaN values. Small amounts of missing data can be removed, while large amounts can be imputed using the mean, median, or mode.

- <b> Handling Categorical Features:</b> Since Scikit-learn only accepts numeric data, categorical variables must be converted using techniques like one-hot encoding, label encoding, or ordinal encoding.

- <b> Outlier Detection:</b> Outliers can skew the model. Methods like Z-score or Interquartile Range (IQR) help detect and manage extreme values.

- <b> Feature Scaling: </b> It is a technique to standardize the independent future present in the data in a fixed range
When features have different ranges, models may bias toward larger values. Scaling techniques such as normalization (min-max scaling) or standardization (mean=0, std=1) ensure all features contribute proportionally.

## Types of Categorical Data

Categorical data represents qualitative values and is divided into two main types:

### 1. Nominal Categories

Nominal data has **no inherent order or relationship** between categories.

**Examples**:

* States
* Branches
* Gender
* Color

Since there is no ranking, assigning numbers directly can mislead the model.

---

### 2. Ordinal Categories

Ordinal data has a **clear order or ranking** among categories, but the difference between them is not necessarily uniform.

**Examples**:

* Grades (A, B, C, D)
* Education level (High School < Bachelor < Master < PhD)

Here, the relative order matters.

---

## Encoding Categorical Data

Most machine learning algorithms in **Scikit-learn** work only with **numerical data**, not strings. Therefore, categorical variables must be converted into numeric form using encoding techniques.

---

## 1. Ordinal Encoding

Ordinal encoding is used **only for ordinal categorical features** where a natural order exists.

### How It Works

Each category is assigned a number based on its order.

**Example**:
Grades:

* A → 4
* B → 3
* C → 2
* D → 1

This preserves the ranking information.

### When to Use

* Input features that have an inherent order
* Example: ratings, education levels, grades

In Scikit-learn:

```python
from sklearn.preprocessing import OrdinalEncoder
```

---

## 2. Label Encoding

Label encoding converts categories into numeric labels.

### Important Clarification ⚠

Label encoding **should not be used for input nominal features** because it introduces an artificial order.

### Correct Use Case

Label encoding is mainly used for the **target/output variable**, especially in classification problems.

**Example** (Target Variable):

* Yes → 1
* No → 0

### Why It Works for Output

The model does not treat target labels as ranked values; it only learns class separation.

In Scikit-learn:

```python
from sklearn.preprocessing import LabelEncoder
```

---

## Key Clarification (Very Important)

| Encoding Method  | Use Case               |
| ---------------- | ---------------------- |
| Ordinal Encoding | Ordered input features |
| Label Encoding   | Target/output variable |
| One-Hot Encoding | Nominal input features |


# 3. One-Hot Encoding

One-Hot Encoding is used when we have <b>nominal categorical data</b> (categories with no inherent order).

It converts each category into a separate binary column (0 or 1), ensuring that no artificial ordering is introduced.

---

## How It Works

Suppose we have a feature:

<b>Color</b> = {Red, Yellow, Blue}

After One-Hot Encoding, it becomes:

<table border="1" style="border-collapse:collapse; text-align:center;">
<tr>
<th>Red</th>
<th>Yellow</th>
<th>Blue</th>
</tr>
<tr>
<td>1</td>
<td>0</td>
<td>0</td>
</tr>
<tr>
<td>0</td>
<td>1</td>
<td>0</td>
</tr>
<tr>
<td>0</td>
<td>0</td>
<td>1</td>
</tr>
</table>

Each category is represented as a <b>binary vector</b>, where:

- 1 → Presence  
- 0 → Absence  

Mathematically:

$$
Category_i \rightarrow [0,0,\dots,1,\dots,0]
$$

This prevents the model from assuming any ordinal relationship.

---

# Dummy Variables

The new binary columns created from one-hot encoding are called <b>dummy variables</b>.

---

# Dummy Variable Trap

Including all dummy variables can introduce <b>multicollinearity</b>, especially in linear models such as:

- Linear Regression  
- Logistic Regression  

---

## Why Does This Happen?

If we have:

$$
n \text{ categories}
$$

Creating:

$$
n \text{ dummy variables}
$$

leads to perfect linear dependency.

Example (3 categories):

If:

Red = 0  
Yellow = 0  

Then:

$$
Blue = 1
$$

So one column can be predicted from others → perfect multicollinearity.

---

## Solution

For:

$$
n \text{ categories}
$$

Keep:

$$
n - 1 \text{ dummy variables}
$$

This is called the <b>n − 1 rule</b>.

In Scikit-learn:

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first')
````

---

# High Cardinality Problem

When a feature has many categories (e.g., 40+ car brands), one-hot encoding:

* Increases dimensionality
* Slows training
* May cause overfitting

---

## Common Solution

* Keep only the most frequent categories
* Combine others into an <b>"Other"</b> category

This reduces dimensionality while preserving key information.

---

# Why Prefer Scikit-learn's OneHotEncoder Over pandas.get_dummies()?

In production ML pipelines, we prefer <b>OneHotEncoder</b> because:

* Integrates with pipelines
* Maintains column order between train and test data
* Ensures consistent transformation
* Handles unseen categories safely

Using <code>get_dummies()</code> may cause issues if new categories appear in test data.

### 2.Feature Construction
Feature construction creates new features from existing ones to capture additional information.

Example: In the Titanic dataset, the columns SibSp (number of siblings/spouses aboard) and Parch (number of parents/children aboard) can be combined into a new feature called Family_Size. This may better reflect social grouping and improve prediction.

### 3.Feature Selection
Feature selection involves choosing only the most meaningful features for model training. Removing irrelevant or redundant features:

- Reduces overfitting

- Improves computational efficiency

- Enhances interpretability

Techniques include correlation analysis, variance thresholding, and model-based feature importance.

### 4.Feature Extraction
Feature extraction algorithmically transforms existing features into new features, often reducing dimensionality while preserving important information.

Example techniques:

- Principal Component Analysis (PCA)

- Linear Discriminant Analysis (LDA)

Feature extraction is often used when the dataset has high dimensionality, helping models focus on the most informative components.
