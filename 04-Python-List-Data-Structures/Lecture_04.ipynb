**List**


**List**:List is a data type where you can store multiple items under 1 name. More technically, lists act like dynamic arrays which means you can add more items on the fly.

Lists are used in programming to store multiple items in a single variable, making data management efficient.
They allow easy access, modification, and iteration over elements.
Lists help reduce code complexity and improve readability.

Array vs List:

Array have static size while list has dynamic size

Array stores elements of the same data type(homogenous); List can store elements of different types(hetrogenous).

Arrays are more memory-efficient and faster for numerical operations; Lists offer more flexibility and built-in methods but have less faster and less memory efficient .

Arrays are available via libraries (like array or NumPy in Python); Lists are built-in and widely used for general-purpose programming.

# **How lists are stored in memory**

Lists in memory are stored as dynamic arrays of pointers (references). Here's how:

Each element in a list is stored at a separate memory location, and the list holds references (addresses) to these elements.

Lists are dynamic — they resize automatically by allocating a new block of memory and copying elements when capacity is exceeded.

This allows fast access by index but may involve overhead when resizing or storing different data types.

# List Comprehension
List Comprehension provides a concise way of creating lists.

newlist = [expression for item in iterable if condition == True]image.png

Advantages of List Comprehension

More time-efficient and space-efficient than loops.

Require fewer lines of code.

Transforms iterative statement into a formula.

```python
# Add 1 to 10 numbers to a list
l=[]
for i in range(1,11):
  l.append(i)
print(l)

#list comprehension
l=[i for i in range(1,11)]
print(l)
```

```python
#scalar multiplication with vectors
v=[3,4,6]
s=-3
x=[]
for i in v:
  x.append(i*s)
print(x)

x=[s*i for i in v]
print(x)
```

```python
l=[1,2,3,4,5]
x=[i**2 for i in l]
print(x)
```

```python
# Print all numbers divisible by 5 in the range of 1 to 50
l=[i for i in range(1,51) if i%5==0]
print(l)
```

```python
# find languages which start with letter p
languages = ['java','Python','php','c','javascript']
[i for i in languages if i.startswith('p') or i.startswith('P')]

# Nested if with List Comprehension
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana','apricot']

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'
[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
# cartesian products -> List comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]

[i*j for i in L1 for j in L2]
```

```python
#print 3x3 matrix using list comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

## Zip

The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.

If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
 # Write a program to add items of 2 lists indexwise

L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]

list(zip(L1,L2))

[i+j for i,j in zip(L1,L2)]

```

```python
# Write a program to check if a list is in ascending order or not
L = [1,2,4,5]

for i in range(len(L)-1):
  if L[i] > L[i+1]:
    flag=0
    break
  else:
    flag=1

if flag==1:
  print("List is in ascending order")
else:
  print("List is not in ascending order")


```
