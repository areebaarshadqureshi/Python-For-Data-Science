Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list, then the 1st index item, and so on till the last element. any leftover items will get added at the end of the new list.

Given List:

list1 = ["M", "na", "i", "Kh"]
list2 = ["y", "me", "s", "an"]
Output:

[['M','y'], ['na', me'], ['i', 's'], ['Kh', 'an']]

```python
list1 = ["M", "na", "i","kh"]
list2 = ["y", "me", "s", "an"]

l=min(len(list1),len(list2))

s=[[list1[i],list2[i]] for i in range(l)]
print(s)

if len(list1)==max(len(list1),len(list2)):
  s.extend(list1[l:])
else:
  s.extend(list2[l:])
print(s)
```

Problem 2: Add new item to list after a specified item
Write a program to add item 7000 after 6000 in the following Python List

list1 = [10, 20, [300, 400, [5000, 6000], 500], 30, 40]
Output:

[10, 20, [300, 400, [5000, 6000, 7000], 500], 30, 40]

```python
list1 = [10, 20,6000, [300, 400, [5000], 500], 30, 40]

for i in range(len(list1)):

  if type(list1[i])==list:
    for j in range(len(list1[i])):
       if list1[i][j]==6000:
            list1[i].insert(j+1,7000)
            break

       if type(list1[i][j])==list:
        for k in range(len(list1[i][j])):
          if list1[i][j][k]==6000:
            list1[i][j].insert(k+1,7000)
            break
  if list1[i]==6000:
            list1.insert(i+1,7000)
            break

print(list1)
```

i.e:

Output:

[1,3,6,10,15,21]

```python
list1 = [1,2,3,4,5,6]
sum=0
l2=[]
for i in list1:
  sum+=i
  l2.append(sum)
print(l2)
```

Problem 11: Write a program that can perform union operation on 2 lists
Example:

Input:

[1,2,3,4,5,1]
[2,3,5,7,8]
Output:

[1,2,3,4,5,7,8]

```python
l1=[1,2,3,4,5,1]
l2=[2,3,5,7,8]
l3=[]

for i in l1:
  if i not in l3:
    l3.append(i)
for i in l2:
  if i not in l3:
    l3.append(i)
print(l3)
```
