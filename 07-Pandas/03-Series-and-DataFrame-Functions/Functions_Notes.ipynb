```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def euc_dist(row):
 p1=row['1st point']
 p2=row['2nd point']

 return np.sqrt((p1[0]-p2[0])**2+(p1[1]-p2[1])**2)

points_df['Euclidean_distance']=points_df.apply(euc_dist,axis=1)
```
