x[-1] gives an error if the index is numeric because pandas treats -1 as a label, not a position.

If -1 isn’t a label in the index, it raises a KeyError.

To access by position (like the last item), use x.iloc[-1].

When the index is non-numeric (like movie names), pandas allows -1 as position safely.

### Copy and Views

- View — points to the same memory as the original. Changing it changes the original too.
- Copy — a completely new independent object. Changing it does not affect the original.

```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# COPY
df2 = df.copy()
df2['A'][0] = 99
print(df['A'][0])   # 1 ← original unchanged ✅

# VIEW (slicing may return a view)
df3 = df[:]
df3['A'][0] = 99
print(df['A'][0])   # may change original ⚠️
```
