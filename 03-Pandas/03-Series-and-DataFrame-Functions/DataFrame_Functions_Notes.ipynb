```python
# find number of movies starting with A for each group
genres['Series_Title'].apply(lambda x:x.str.startswith('A').sum())
```

```python
def foo(group):
  return group['Series_Title'].str.startswith('A').sum()

genres.apply(foo)
```

```python
filter=movies[movies['Series_Title'].str.startswith('A')]
filter.groupby('Genre').count()
```

```python
# find ranking of each movie in the group according to IMDB score
movies['Rank_movies'] = genres['IMDB_Rating'].rank(ascending=False, method='first')
```

```python
genres.get_group('Action')
```

```python
def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group

genres.apply(rank_movie)
```

```python
# find normalized IMDB rating group wise
def foo(group):
 X_min=group['IMDB_Rating'].min()
 X_max=group['IMDB_Rating'].max()
 group['normalized_Rating']=(group['IMDB_Rating']-X_min)/(X_max-X_min)
 return group

genres.apply(foo)
```

```python
# groupby on multiple cols
duo = movies.groupby(['Director','Star1'])
duo
# size
duo.size()
# get_group
duo.get_group(('Aamir Khan','Amole Gupte'))
```

```python
# find the most earning actor->director combo
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# find the best(in-terms of metascore(avg)) actor->genre combo
best=movies.groupby(['Genre','Star1'])
best['Metascore'].mean().sort_values(ascending=False).head(1)
```

```python
best['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(1)
```

```python
# agg on multiple groupby
duo.agg(
    {
        'Gross':'sum',
        'IMDB_Rating':'mean'
    }
)
```

```python
ipl = pd.read_csv('/content/deliveries.csv')
ipl.head()
```

```python
ipl.info()
```

```python
# find the top 10 batsman in terms of runs
match=ipl.groupby('batsman')
match['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
r=ipl[ipl['batsman_runs']==6]
r.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
last_over=ipl[ipl['over']>15]
six_four=last_over[(last_over['batsman_runs']==4)|
 (last_over['batsman_runs']==6)]

six_four.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
# find V Kohli's record against all teams
vk=ipl[ipl['batsman']=='V Kohli']
vk.groupby('bowling_team')['batsman_runs'].sum()
```

```python
# Create a function that can return the highest score of any batsman
def foo(batsman_name):
  vk=ipl[ipl['batsman']==batsman_name]
  return vk.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1)

foo('CH Gayle')
```
