### `Q-4:` Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
def z_normalization(group,cols_to_perform):
  for col in cols_to_perform:
    group[col]=(group[col]-group[col].mean())/group[col].std()
  return group

cols_to_perform=['Passes','Passes Completed','Attempted Line Breaks','Completed Line Breaks']
fifa_df.groupby('Team').apply(lambda group: z_normalization(group, cols_to_perform))
````

###`Q-5:` Find batsman in below category-
* Highest score while chasing
* Best Strike rate while chasing and have faced 100+ balls


> Chasing mean team batting in second inning

````python
ipl.info()
````

````python
chasing=ipl[(ipl['innings']==2)]
balls_faced=chasing.groupby(['ID','batter']).size()
eligible_batters = balls_faced.groupby('batter').sum()
eligible_batters =eligible_batters [eligible_batters >=100].index
chasing_batsman=chasing[chasing['batter'].isin(eligible_batters)]
runs=chasing_batsman.groupby('batter')['batsman_run'].sum()
final=runs.sort_values(ascending=False).head(1)
balls=chasing_batsman.groupby('batter').size()
strike_rate=round((runs/balls)*100,2).sort_values(ascending=False).head(1)
print(f'Best Batsman who chased more runs:\n{final}')
print("🎯 Best Strike Rate while Chasing (Min 100 Balls Faced):")
print(strike_rate)
````

````python
combo=ipl.groupby(['bowler','batter'])
combo['player_out'].count().sort_values(ascending=False).head(1)
````

````python
# Step 1: Filter rows where player_out is not null (i.e., actual dismissals)
dismissals = ipl[ipl['player_out'].notnull()]

# Step 2: Count dismissals per (bowler, batter) pair
dismissal_counts = dismissals.groupby(['bowler', 'batter'])['player_out'].count()

# Step 3: Sort and get the top pair
most_successful_pair = dismissal_counts.sort_values(ascending=False).head(1)
print(most_successful_pair)
````

###`Q-8:` Make a dataframe for all batting pairs played together.
```
Batsman1 Batsman2 Runs Avg StrikeRate
```

> Just to ease this question you can count wide-balls for strike rate.

````python
pairs=ipl.groupby(['batter','non-striker'])['batsman_run'].agg(['sum','mean']).reset_index()
pairs.columns = ['batter', 'non-striker', 'Total_Runs', 'Average_Runs']

pairs_data=pd.DataFrame(pairs)
pairs_data
````

````python
pairs_data.sort_values(by='Total_Runs',ascending=False,inplace=True)
pairs_data
````

````python
ball_faced=ipl.groupby(['batter','non-striker']).size()
pairs_data['Ball_faced'] = pairs_data.apply(lambda row: ball_faced.get((row['batter'], row['non-striker']), 0), axis=1)
pairs_data
````

````python
pairs_data['Strike_rate']=round(pairs_data['Total_Runs']/pairs_data['Ball_faced']*100,2)
pairs_data.rename(columns={'batter':'Batsman 1','non-striker':'Batsman 2'},inplace=True)
pairs_data
pairs_data.set_index('Batsman 1')
````
