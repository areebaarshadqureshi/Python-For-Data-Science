**Nested loop**:
A nested loop is a loop inside another loop.
In Python, this means the inner loop runs completely for each iteration of the outer loop.
Nested loops are often used to work with 2D data, like grids, matrices, or tables.

```python
#Find unique pairs
n=int(input("Enter n:"))
m=int(input("Enter m:"))
for i in range(1,n+1):
  for j in range(1,m+1):
       print(f"({i},{j})",end=",")
  print("\n")
```

```python
n=int(input("Enter n:"))

for i in range(1,n+1):
  for j in range(1,i+1):
     print("*",end="")
  print()
```

```python
#pattern 2
n=int(input("Enter n:"))

for i in range(1,n+1):
  for j in range (1,i+1):
     print(j,end="")
  for k in range(j-1,0,-1):
     print(k,end="")
  print()
```

**Loop Control Statement:**

**Break:** Used to exit a loop immediately.

**COntinue:** Skips the current iteration and goes to the next one

**Pass:** Does nothing — it's a placeholder when code is required syntactically but you don’t want to write anything yet.

```python
#Break
for i in range(1,11):
  if i == 7:
    break
  print(i)
```

```python
#You have a range of numbers between 10 and 100 and you have to find all prime numbers between this range
#prime number has only two factors
for i in range(10,101):
  for j in range(2,i):
    if i%j==0:
      break
  else:
    print(i)


```

```python
#continue

for i in range(1,11):
  if i == 7:
    continue
  print(i)
```

```python
#pass
for i in range(1,11):
  if i == 7:
    pass
  else:
    print(i)
```

**Difference between unicode and ascii characters**

ASCII is old and limited to English characters like A–Z, a–z, 0–9, symbols.
It is 8 bits size (2^8)

Unicode is modern and supports all languages and emojis.
It has 8,16 and 128 bit size (2^16)

**Accessing Substrings from a string**

**Indexing"":Indexing means accessing individual characters in a string (or elements in a list) using their position.


```python
#positive indexing (0 to end)
s='Hello world'
print(s[4])

#Negative indexing (-1 to -n)
print(s[-4])
```

```python
#slicing:Slicing is used to extract a part of a string or list using [start:stop].

#example:

s="Hello World"
print(s[1:4])
print(s[1:])
print(s[:4])
print(s[:])
print(s[::2])
print(s[6:0:-2])
print(s[::-1])
```

```python
#Negative indexing example
print(s[-5:])
print(s[-5:-2])
print(s[:-5])
```

```python
#Check Palindrom

n=input("Enter any string:")
j=0
for i in range(len(n)-1,-1,-1):
  if n[j]!=n[i]:
     flag=0
     break
  else:
     flag=1
     j+=1

if flag==1:
  print(f"{n} is a palindrom")
else:
  print(f"{n} is not a palindrom")
```

```python
#title case without using the title()

n=input("Enter any string")
s=""
m_upper=True

for i in n:
  if m_upper:
    s+=i.upper()
    m_upper=False
  elif i==" ":
    s+=i
    m_upper=True
  else:
    s+=i

print(f"Sent:{s}")
```
